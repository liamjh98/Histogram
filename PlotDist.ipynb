```python
from numpy import *;
import seaborn as sns;
import pandas as pd;
import numpy as np;
import matplotlib.pyplot as plt;
import statistics;
sns.set_theme();
sns.set(style="darkgrid");
import subprocess, os, platform

x1 = loadtxt('C:/Users/peppe/Desktop/MXP.txt',dtype=float,usecols=(0,));
x2 = loadtxt('C:/Users/peppe/Desktop/MX_V.txt',dtype=float,usecols=(0,));
x3 = loadtxt('C:/Users/peppe/Desktop/M23C6_Cr.txt',dtype=float,usecols=(0,));

df = pd.concat(axis=0, ignore_index=True, objs=[
    pd.DataFrame.from_dict({'value': x1, 'Precipitate': 'Nb'}),
    pd.DataFrame.from_dict({'value': x2, 'Precipitate': 'V'}),
    pd.DataFrame.from_dict({'value': x3, 'Precipitate': 'Cr'})
]);

kwargs = dict(kde=True, binwidth=6, kde_kws=dict(cut=3), alpha=.6, edgecolor=(1, 1, 1, .4), palette=['b', 'g', 'c']);

fig, ax = plt.subplots(figsize=(16,9), dpi= 80);
sns.histplot(
    data=df, x='value', hue='Precipitate', multiple='dodge', ax=ax, **kwargs
);
ax.set_xlim([0, 300]);

font = 35
plt.title('Size of precipitates in G91',fontsize=font);
plt.xlabel('Size [nm]',fontsize=font);
plt.ylabel('Number of precipitates',fontsize=font);
plt.yticks(fontsize=font)
plt.xticks(fontsize=font)
plt.setp(ax.get_legend().get_texts(), fontsize='35') # for legend text
plt.setp(ax.get_legend().get_title(), fontsize='35') # for legend title
#sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))

file = plt.savefig("C:/Users/peppe/Desktop/Size_G91_Precipitates.pdf", transparent=False, bbox_inches='tight');
os.startfile("C:/Users/peppe/Desktop/Size_G91_Precipitates.pdf", 'open')
```
